```python
import argparse
from PIL import Image
from timeit import default_timer as timer
import test
import pandas as pd
import numpy as np
import random
import keras
import matplotlib.pyplot as plt 
```

```python
final_data = np.load('Data/Model_Weights/model_history_final.npy', allow_pickle=True)
initial_data = np.load('Data/Model_Weights/model_history_stage_1.npy', allow_pickle=True)
```

```python
loss_values = initial_data[()]['loss']+final_data[()]['loss']
val_loss_values = initial_data[()]['val_loss']+final_data[()]['val_loss']
# mae_values = initial_data[()]['mean_absolute_error']+final_data[()]['mean_absolute_error']
# val_mae_values = initial_data[()]['val_mean_absolute_error']+final_data[()]['val_mean_absolute_error']
```

```python
epochs = 40
plt.plot(loss_values, 'bo', label='Training loss')
plt.plot(val_loss_values, 'b', label='Validation loss')
plt.axis([1,40,0,70])
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.savefig('losses.png', bbox_inches='tight')
plt.show()
```
